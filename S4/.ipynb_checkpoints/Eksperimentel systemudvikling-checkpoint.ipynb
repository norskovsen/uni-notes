# A fusion of skills
- The key concerns of a designer of an interactive system is
    - Design
    - Technologies
    - People
    - Activities and contexts
    
    
- It is useful to distinguish the amount of formality associated with design
    - At one end of the spectrum is **engineering design** where scientific and technical principles are used 
    - At the other end of the spectrum is creative or artistic design
    - In the middle is **design as a craft** that draws upon engineering and creative methods

## Interface
- The **interface** to an interactive system is all those parts with which people come into contact, physical, perceptual and conceptual 
    - Physically we might interact using buttons or moving levers and the system might respond with feedback through these 
    - Perceptual the device might display things on the screen or things we can hear
    - Conceptually we interact by the system by figuring out what is does and what it should be doing. The device provides messages and other displays with helps us figuring that out

## Being human-centered 
- Being human-centered is about
    - Thinking about what people want to do rather that what technology can do 
    - Designing new ways to connect people with people
    - Involving people in the design process
    - Designing for diversity 
    
## Information appliances
- Information appliances has the following characteristics 
    - Appliances should be everyday things that only require everyday skills to use
    - Appliances should have a clear, focused function that can be used in a variety of circumstances 
    - Peer-to-peer interaction: appliances work together without the need for central control of uploading and downloading 
    - Direct user interface: they should be simple and easy to use
    - Successful appliances are those which support a swift and simple completion of tasks
    - Represents the idea of doing something on impulse without having to think much about it
    - Are personal and portable

## People 
- People differ in physical characteristics such as height and weight 
- The differences in the five senses has a huge effect on
    - how accessible a device is
    - how usable it is 
    - how enjoyable it is to use 
___

- The term **ergonomics** describes the study of the relationship between people and the environment
    - The best known example of ergonomic knowledge is **Fitts' Law**
    
- **Anthropometrics** means the measure of man
    - Figures have been compiled from thousand of measurements on different people
    - can e.g. be used to tell the designer if and average person can press two buttons at the same time
    
### Fitts' law    
- **Fitts' Law** is a mathematical formula which relates the time required to move a target as a function of the distance to the target and the size of the target itself. It is expressed as follows
$$
T_{\text{time to move}} = k \cdot log_2 (D/S + 0.5)
$$
    where $k \sim 100 \text{ms}$, $D$ is the distance between the current cursor position and the target and $S$ is the size of the target     

### Psychological differences
- Designers should be able to design for people with poor ability for by providing good signage and clear direction
- Language differences and cultural differences effect how people interpret things
- People have different needs and abilities when it comes to attention and memory 
    - These can change based on stress and tiredness
    - People are better at recognizing than remembering things 
- People have different experiences and therefore have different conceptual models

### Mental models
- A **mental model** is the understanding and and knowledge that we posses of something 
    - If people don't have a great mental model of things they can only perform action by rote and won't be able to recover if something goes wrong
- A key design principle is to design things so that people will form a correct and useful mental models of how the work and what they do
    - It is important for design to provide sufficient information in the interface for people to form an accurate mental model 
- People interact with the **system image** and from this they derive their conception 
![system_image](img/system_image.png)
- **Normans observation about the nature of mental models**
    - Mental models are incomplete: People will understand some parts of the system better than others
    - People can try out their models when required, but often with limited accuracy 
    - Mental models are unstable: people forget things
    - Mental models don't have firm boundaries: similar devices and operations get confused with one another
    - Mental models are unscientific, exhibiting superstitious behaviour.
    - Mental models are parsimonious: people are willing to take additional physical operations to minimize mental efforts e.g. restarting a computer

## Activities 
- The term activity is both used for very simple tasks as well as highly complex 


- The main features of an activity is
    - Temporal aspects
    - Cooperation
    - Complexity
    - Safety-critical
    - The nature of the content

### Temporal
- **Temporal** aspects of an activity cover how regular or infrequent an activity is
    - Designer should ensure that frequent tasks are easy to do
    - Designers should ensure that infrequent task are easy to learn (or remember) how to do
    
    
- Important features of activities also include time pressures, peaks and troughs of working.
    - A design that works well when quiet can be awful when things are busy 
    
    
- Some activities take place as a single continuous set of action where as others are more likely to be interrupted 
    - If people are interrupted doing and activity the design must ensure that they easily can *"find their place"* again
    - It is important to ensure that people don't make mistakes and or leave out important steps
    
    
- It is important to think about the **response time** of an activity
    - People expect a response time of 
        - about 100 milliseconds for **hand-eye coordination** activities
        - 1 second for **cause-effect relationship** such as clicking a button 
    - Anything more than five seconds and people feel frustrated and confused
   

### Cooperation
- It is and important feature to know if the operation is carried out alone or concerned with working with others
- Issues of awareness of others and communication and coordination become important when working with others

### Complexity
- Well defined task need different designs than vague tasks
    - If a task or activity is **well defined** it can be accomplished using a simple step by step design 
    - A **vague activity** means that people have to be able to browse around, see different types of information and move from one thing to another and so on

### Safety critical
- Some activity are **safety critical** in which case any mistake can result in an injury or serious accident others less so
    - Where every safety is involved designers must ensure that mistakes do not have a serious effect 

### The nature of content
- It is also important to think about the data requirements for the activity 
    - If large amount of data have to be input as part of the activity then a keyboard is certainly needed
    - Some activity may need large graphic display to display videos and photos 

## Context
- Activities always happens in a context

    - The **organizational context**
    - The **social context**
    - The **physical circumstances**

### Physical environment
- The physical environment in which and activity happens is important 
    - e.g. the sun shinning on an ATM display
- The environment may be noisy, cold, wet or dirty 

### Social context
- The social context which an activity takes place is important
- A supportive environment will offer plenty of help for the activity
- There maybe privacy issues to consider
- Interaction maybe very different when people are alone vs when they are not
- Social norms may dictate the acceptability of certain designs 

### Organizational context
- Organizational context is important because changes in technology often alter communication and power structures 

## Technologies

### Input

- Input devices includes
    - Buttons or switches
    - Different types of mouse 
    - Touch screens
    - Motion or other sensor controls
    - Joysticks 
    - Keyboards

### Output 
- Technology for displaying content to people rely primarily on the three perceptual abilities of vision, hearing and touch
- Sound is a very important type of output
- Haptics allow us to be in touch with interactive devices that is direct and immediate 

### Communication 
- When designing communication between devices and people bandwidth and speed are critical 
- Feedback is important so they know what is going on

### Content 
- Content concerns the data in the system and the form it takes
- Good content is accurate, up to date, relevant and well represented 
- In some technologies content is everything, in others are concerned with features and some are a mixture of both 

## Scoping a problem with PACT
- The aim of human-centered interactive system is to achieve the best combination of PACT elements with respect to the domain
- A PACT analysis is useful for both analysis and design activities 
- To do a PACT analysis the designer scopes out the variety of Ps, As, Cs, and Ts that is possible or likely in a domain
    - This can be done by brainstorming and other techniques and by working with people through observations, interviews and workshops 
    - A PACT analysis is also useful for developing personas and scenarios
    - The designer should look at trade offs between the PACT elements
___

- For **people**, designers need to think about physical, psychological and social differences and how the differences changes in certain circumstances and over time
    - It is most important that designers think about the different stakeholders in a project
- For **activities**, designers need to think about
    - **The complexity of the activity:** focused or vague, simple or difficult, few steps or many
    - **The temporal features:** frequency, peaks and troughs, continuous or interruptible
    - **Cooperative features** 
    - **The nature of data**
- For **context**, designers need to think about physical, social and organizational setting 
- For **technology**, designers need to concentrate on input, output, communication and context

## Understanding
- **Understanding** is concerned with 
    - what the system has to do
    - what it is like
    - how it has to fit in with other things
- Designers need to research people, activities, contexts relevant to the domain
    - They need to understand the opportunities and constraints provided by technologies
- There are both functional and non-functional requirements to consider
    - Functional requirements are concerned with what the system should be able to do and with the functional constraints
- There are also logical and organizational constraints which makes some designs infeasible
- Requirements are generated through discussion and interactions with people who will use or be affect by the system 
    - are also generated through observations of existing systems, what people like and do not like
    - can also be generated using focus groups
- **Stakeholders** are all the people who will be affected by the proposed design

## Design 
- Design activities concern both conceptual design and physical design 

### Conceptual design
- **Conceptual design** is about considering what information and functions are needed for the system and its purpose 
    - Its about deciding what a person needs to know to use the system
    - It is about finding a clear conceptualization to people so that they quickly develop a clear mental model 
- One way to conceptualize the main features of a system is to use a **rich picture**
    - Captures the main conceptual relationships between the main conceptual entities in a system 
- The key feature of conceptual design is to keep things abstract
    - focus on the what rather than the how 
- There are no clear cut between conceptual and physical design 

### Physical design 
- Physical design in concerned with how things are going to work and detailing the look and feel of the product
- There are three components to physical design 
    - **Operational design:** is concerned with specifying how everything works and how content is structured and stored
    - **Representational design:** is concerned with fixing the color, shapes and size and information layout 
    - **Design of interactions:** is concerned with allocation of function to humans or to technology     

## Envisionment
- Is concerned with finding the appropriate media to render design ideas
- Can be e.g.
    - sketches 
    - fully functioning prototypes
    - cardboard mock-ups

## Scenarios throughout design 
- Scenarios are useful for understanding, envisioning, evaluation and both conceptual and concrete design
- There are four different types of scenarios 
    - **Stories** are real world experiences of people   
    - **Conceptual scenarios** are more abstract descriptions in which some details have been striped away 
        - Abstracts the observed problems from the stories
    - **Concrete scenarios** are generated from abstract scenarios by adding specific design decisions and technologies 
        - One conceptual scenario generates lots of concrete scenarios
        - Useful for 
            - prototyping 
            - envisioning design ideas
            - evaluation
        - Not a clean break between conceptual and concrete 
    - **Use cases** describes the interaction between people and devices
        - How the system is used
        - A set of use cases specifies the functionality of the system and the interactions

# Understanding 
## Requirements
- A **requirement** is something a product must do or a quality it must have
- Requirements are found by the designer by studying the current activities and stories
- Evolve over time in the iterative analysis/design/evaluation process
- There are two types of requirements 
    - **Functional requirements** are what the system must do 
    - **Nonfunctional requirements:** are a quality that the system must have 
        - The way the functionality operates
        - Data requirements are also important
- When suppling requirements supporting evidence is a good idea to include
    - such as e.g. interviews interview, observation, reports
    - helps the reader understand the reasons
- It is important to use a **template** for requirements which should at minimum contain
    - A unique reference number 
    - A one-sentence summary 
    - The source(s) of the requirement
    - The rational for it 
- Requirements should be reviewed with customers and clients and modified as necessary 
- Decisions should be made about the priority of the requirements this can be done using the **MoSCoW** rules which classify requirements into 
    - **M**ust have - fundamental requirements without them the system would not be usable 
    - **S**hould have - would be essential if more time was available
    - **C**ould have - of lesser importance, can easily be left out of current development
    - **W**ant to have bu **W**on't have this time around - can wait till latter development



## Interviews
- Interviews with all the various stakeholders in the domain are vital for gathering stories
- Designers use a wide range of interviews
    - **Structured interviews** uses question that are developed beforehand
        - are easy to carry out because of pre-structuring
        - people are limited to restricted replies
        - difficult for the interviewer to follow up on an unexpected response
    - **Semi-structured interviews** are where the interviewer is prepared with questions but can reword these appropriate
        - very frequently used
        - interviewer often prepares a a checklist
        - more demanding for interviewer
        - higher data quality
    - **Unstructured interviews** are where there are no preset questions or topics beyond the general subject of the project in question 
- Interviews are a good source of stories and are good for discussing early prototypes


- **Think aloud commentaries** is where users are asked to talk through the operations
    - useful for information about low-level detail about the current technology
    - data is called 'verbal protocol'
    - can provide helpful indication of current problems
    - not useful for all problems 
    
    
- Practical considerations in interviewing 
    - **Preparation**
        - Get to know the background information
        - Idiot questions can uncover unspoken assumptions
        - For work activities background research might include e.g. company reports, brochures, websites and organization charts
    - **Keeping track of the interview**
        - Being a pair of interviews can be used, so that one takes notes and the other interviews
        - Note-taken can be relieved using audio or video recording
            - Check equipment periodically 
            - Notes are still useful such as time 
            - Notes are vital if the recording has gone bad
    - **Telling stories**
        - Stories can be misleading
        - It is important to remember that people can be giving such things disproportionate emphasis 
        - Designer are listening for current problems
    - **Reflection and exploration**
        - It is a good idea to have the interviewer read the summery of the interview
        - Look at notes to find point that need clarification
    - **General-purpose exploratory questions**
        - Are useful in early stages or taciturn interviewee
        - Some useful are 
            - *'Tell me about your typical day.'*
            - *'Tell me three good things about ...'*
            - *'and three bad things.*
            - *'What if you had three wishes to make the application better?'*
            - *'What has gone wrong with the application recently? How did you cope?'*
            - *'What else should we have asked about?'*
    - **When to stop**
        - Practical constraints against the comprehensiveness of the data
        - All significant stakeholder groups must be covered 
            - Two or three interviewees per role / group across four different types of organization 
        - General rule is to stop when no new insights are being obtained

## Questionnaires 
- **Questionnaires** are suited to gathering a large amount of quantifiable data
- Questionnaires need to be designed, prototyped and evaluated
- For a small number of people a interview will obtain the same information in a more manageable way    
- A good questionnaire 
    - are understandable
    - are unambiguous
    - collect data which actually answers evaluation questions
    - can be analysed easily.
- Low response rates
- Most people finish the questionnaires when they are conducted in person but not one the web
- The Likert scale are used to indicate peoples agreement with a statement using a five point scale 
    - Strongly agree
    - Agree
    - Neutral
    - Disagree
    - Strongly disagree
- It is hard to get the wording right so the information you are getting are useful


- Hints and tips 
    - **Specific questions are better than general ones**
    - **Closed questions are usually preferable to open questions**
        - helps avoid differences in interpretation
        - open questions are useful for seeking comments in respondents own words
    - **Consider a no-opinion option**
        - avoid that people have to manufacture option
        - a middle opinion can also be used
    - **Vary the orientation of rating scales or intersperse with other questions**
        - avoid people selecting the same opinion
        - reverse some scales or add different types of question in between
    - **Appearance, order and instructions are vital**
        - should look easy to fill in 
            - initial questions should be easy and interesting 
            - middle questions cover more difficult areas
            - last questions should be interesting to encourage completion
        - keep design simple 
        - give clear instructions
        - ticking boxes is easier than circling answers
    - **Add introductory and concluding notes**
        - the introduction should 
            - explain the purpose
            - assure confidentiality 
            - encourage reply
        - the concluding notes should
            - ask respondents to check that they have answered all questions 
            - encourage early return, if there is a deadline 
            - offer to send summery of findings, if appropriate 
            - thank them for help
    - **Make return easy**
        - using internal mail
        - a box to be placed in a convenient place 
        - web can be used 
        - post returns should include prepaid envelope

## Card sorting techniques
- **Card sorting** refers to a number of techniques concerned with understanding how people classify and categorize things
    - the basic method involves writing concepts onto cards and then grouping them in different ways
    - can be done face-to-face or using an online tool 
    - can be a very effective method of gaining insight into how people think about things and classify them 
- There are two types of card sort
    1. An open card sort 
        - Starts with blank cards
        - Participants are asked to write down the objects or actions they think are important in some domain
        - The cards are then gathered together into categories
    2. A closed card sort
        - Starts with predefined categories and asks participants to place objects into the categories

# Contextual Design 1
- **Contextual Design** is a practical design method that focuses on fitting software and hardware solutions to its environment
- The designer is seen as a *detective*, someone who is attempting to make sense of a real-life working situation
- In CD **customer** is anyone who uses the system in question either directly or indirectly
    - Similar to th term 'stakeholder'
- CD supports the designers in data collecting    
- CD is intended to be used by a design team 

## Contextual Inquiry
- Contextual Inquiry (CI) is the first stage of CD 
    - is essentially a combination of a focused interview and observation
    - is made up from a number of techniques that are designed to help the analyst understand what it is a client needs.
    - brings together a number of techniques including interviewing, artefact collection and observation under one unifying theme or philosophy


- The are four guiding principles of Contextual Inquiry
    - **Context:** Go to the customer's workplace and observe how work is actually carried out
        - Allows the analyst to experience everyday detail of work
        - May need to be supplemented by retrospective explanations
    - **Partnership:** The analyst and customer are expect in different fields
        - The analyst should be looking for patterns and structure in the work
        - The customer contributes her knowledge of how the work really gets done
        - The customers can influence the analyst's interpretations
    - **Interpretation:** The analyst must interpret workplace data so it is properly understood
        - The analyst should reflect her interpretation back to the customer
    - **Focus:** Each site visit and interview needs a focus
        - Concentrating on one part of the work helps see detail be at the expense of other aspects
        - Can be helped by sharing data in group 
        - A sign of over-narrow focus might be unexpected actions by the customer in a interview
        
        
### Practicalities
- It is essential to have a notebook and ideally a tape-recorder or video-camera
- The first 15 minutes or so should take care of 
    - usual introductions
    - obtaining of permissions 
    - explanations of the process
    - focus of the interview
- The Contextual interview then follows
    - May be around 2-3 hours' mixed conversation and observation
    - The interviewee may move around the workplace
    - Try to observe as many different types of everyday tasks as possible
    - To observer as many different types of everyday tasks as possible, the following suggestions can be used
        - **Intermittent, unscheduled and brief tasks** e.g. recovering from a system crash:
            - Have customers note the event, save any related artefacts, then discuss in a follow up interview
        - **Uninterruptible work** e.g. training session or client meeting:
            - Videotape or take detailed notes
            - Review later with the customer
        - **Very length collaborative projects** e.g. major engineering design projects:
            - Interview different customers at different stages in the process
            - Collect project documents and other artefacts and walk through the process with customers
        - **Focused, fine-grained tasks** e.g. interacting with a complex piece of software:
            - Videotape and review later with customer, use with caution
- Use the last 15 minutes of the interview to review what you have learned
    - Remember to ask your interviewee what you might have missed 

## Contextual Design: Work modelling 
- The next stage is to document the raw data obtained from interviews and observation as a set of models
    - The models each represents a different facet of work

### Flow Models
- A **flow model** represents how work is broken up across people and how it is coordinated to get the whole job done
    - It is drawn from a particular point of view
    - There are several, one for each point of view


- Components of a flow model
    - **Individuals:** Who is involved
        - Thinking about the individuals themselves rather than their roles
        - They are labelled with an identifier such as 'customer 1' and job title rather than personal names
    - **Responsibilities:** belonging to each individual 
        - e.g. booking rooms
    - **Groups:** is more than one person with the same responsibilities 
        - They can be considered a group if the outside world interacts with all its members in a similar way
        - e.g. receptionists
    - **Flow:** How people communicate to get the work done
        - Informal chat or formal artefacts such as invoices
    - **Artefacts:** the things that are processed or support work
        - can be tangible objects such as members or intangible such as important conversation where key points are agreed upon 
    - **The communication topic or action:** such as request for a room
    - **Places:** such as meeting rooms
        - are shown if they are central to coordinating the work
    - **Breakdowns:** problems in communication or coordination     
    

- Hints and tips for creating flow models
    - Represent every contact people make.
    - Note what responsibilities people take on
        - even those that are not part of their job.
    - Look at the action people take without thinking.
    - Represent locations, things and systems when they make a place to coordinate.
    - The real interactions between people reveal glitches in the work.

### Sequence Models
- A **sequence model** represents work tasks
    - are ordered and unfold per time
    - can be shown as a sequence of steps of actions
    - are drawn from a particular point of view
    - are likely to have several different version for some tasks at least 
    

- Components of a sequence model
    - The **Intent** (or purpose) that the sequence is intended to achive
        - There will always be one main intent
        - Usually several secondary intents
    - The **trigger** that cause the sequence of action
        - The main intent and the trigger are a pair
        - The intent is the reason why a sequence action is taken
        - The trigger is the event that sets it off
        - There is only one trigger for each sequence
    - A series of **steps** that achieve the intent
        - There should be enough detail to allow a colleague to implement software to support the activity without the need for help
    - Any **breakdowns** or problems


- Hint and tips
    - Capture actions at the level that matters for your project.
    - Customers’ actions are never purposeless.
    - Watch how automation removes effective prompts to action.
    - Find intents implied by actions.   

### Artefact Models
- Artefacts are the things people use or create in their work
    - are collect under CI 
- Often the interface to the computer system is a key artefact
- The **arteface model** consists of the artefact or a sketch photo photocopy or scanned version of it, labeled to indicate the aspects of interest 
    - The model has at least two uses
        - It can be used to tease out detail of how the work is done currently, preferably with customer
        - Later in the design process it provides basic information as 
            - to what data or other material is currently held/processed/communicated
            - its current organization and problems occurring
    - Sometimes the current artefacts can be used as a basis for the data structures for the new system
    
            
- Components of the artefact model
    - The **information** content itself
        - e.g. the rooms booked and guest details in the booking diary
    - The **structure** of the object into different parts
        - showing different usages, who gets involved and the information intended for different users
    - Informal **annotation** of the artefacts
        - e.g. the use of Post-it notes and handwritten notes
    - The **presentation** of the artefact
        - general style, the use of color and fonts
        - contains clues both to organizational image and values and to the relative importance of different parts of the artefact
    - Note any aspects that change over **time**
    - Note **when** it was created, **what** is is used for, and **by whom**
    - Any **breakdowns** in its use
    
- Hints and tips 
    - Artefacts capture traces of people's work practice.
    - Structure reveals how the work is organized.
    - Content is the trail left by real events.
    - Presentation directs the eye and reveals importance.

### Cultural model
- Work takes place in a culture that defines people's expectations, wishes, organizational policies and values. 


- Components of the cultural model
    - **Influencers** people or bodies which affect the way work is done
        - Typically managers, customers or competitors of the organization being studied
        - shown as bubbles
    - The **extent**  to which work is affected by the influencers 
        - indicated by the degree of overlap between bubbles
    - The **direction** of the influence, shown as arrows
    - **Breakdowns** arising from cultural issues.
    
    
- Hints and topics
    - The cultural model speaks the words that people think but don't say
    - An organization's culture is not reflected in its organization chart

### Physical Model
- The **physical model** is a representation of where the work takes place
    - not an exact floor plan
    - shows the key features of the workplace
    - helps to show why work is carried out in a particular way
    - also indicates physical features that may need an equivalent in a computer system


- Components
    - The **physical structures** of the works
        - They affect the way work is (or could be) carried out
    - **Movement** within the workplace 
        - The paths of people's regular movements
        - How significant artefacts are moved around
    - **Communications** and **computer technologies** and **networks**
    - The location of **key artefacts** created or processed
    - The **layout** of the workplace
    - **Breakdowns**

# Envisionment
- **Envisionment** is fundamental to effective human-centered design
    - Enables designers to see things from other peoples perspectives 
    - Can be used in many different situation to communicate design
- Envisionment is about bringing abstract ideas to life 
    - Helps exposing flaws and difficulties 
___
    
- Suggested series of steps for the envisionment process:
    1. Review requirements and conceptual scenarios 
    2. Develop representation of your design ideas 
        - Should at minimum include 
            - Concrete scenarios 
            - Storyboards for developing main interaction sequences
            - Snapshot sketches of key scenes and other aspects of the product
    3. If the product is a new one experiment with different metaphors and design concepts through representations
    4. Explore design ideas with people who will be using the system whenever possible 
    5. Develop wireframes to provide more detail on the proposed structure and navigation
    6. Iterate and gradually formalize the design through prototypes and further evaluations

## Basic Techniques
- **Sketches and snapshots**
    - Sketching allows a designer to quickly visualize their ideas and thoughts 
    - Individual snapshots can be provided to show key moments of interaction 
        - are useful for exploring the impact of a design 
        - can be single sketches or frames from a storyboard  or can be produced using software
- **Storyboards**
    - A simple cartoon like structure
    - Key moments from an interactive experience is represented 
    - Allows you to get a felling of the flow of the experience
    - There are three main types of storyboarding
        - **Traditional storyboarding:** like a storyboard from a film with some notes attached
        - **Scored storyboards:** If the application has a lot of motion graphics information about them can be attached to the storyboard e.g. type, colors, images and/or sound
        - **Text-only storyboards:** useful if the application has a lot of complex sequences where you can specify what images appear, what text accompanies them, any accompanying media, general notes about tone, flow, etc. 
- **Mood boards**
    - You gather visual stimuli that capture something you feel about the design 
        - such as photographs and other images, colors, textures, shapes, headlines from newspapers or magazines, quotations from people, pieces of fabric and so on.
    - Attach the stimuli to a pinboard
    - The point of the board is to act a inspiration
    - Getting a client to make a mood board gives the designer some insight into the kinds of aesthetics that appeal to them
- **Navigation maps**
    - Focus on how people moves through the system 
    - Each page or location is represented with a box or heading
    - Every page that can be accessed from given page should flow from it
    - Can be used in scenarios 
![navigation_maps](img/navigation_maps.png)    
- **Wireframes** 
    - are outlines of the structure of a software system 
    - focuses on the structure of particular types of pages

## Prototypes
- It is important that a prototype is interactive
- Is especially useful for clients to capture the outcome of the envisioning techniques
- The prototype might seek to highlight just the interface, or some crucial aspect of the functionality
- A **horizontal prototype** aims to go across the whole system, but deals only with top-level functions so much of the detail is omitted
- A **vertical prototype** implements a full range of features, from top to bottom, but is only applied to a small number of functions of the overall system 
- Different types of prototypes:
    - **Hi-fi prototypes**
        - are similar in look and feel to the final product, not necessarily in functionality
        - produced in software 
        - is useful of detailed evaluation of main design elements 
        - often crucial for client acceptance 
        - developed fairly well into the project when the ideas begin to firm up 
        - accurate detail is vital
    - **Lo-fi prototypes**
        - often termed paper prototypes
        - are more focused on the broad underlying design ideas
        - are designed to be produced quickly 
        - capture early design thinking
        - aid the process of generating and evaluating many design solution
        - is a series of screenshots that people can walk through 
        - main issues:
            - **Robustness:** if handled by lots of people it needs to be though enough to survive
            - **Scope:** focus on broad issues and key elements
            - **Instructions:** trade-off between adding enough detail and too much detail
            - **Flexibility:** parts of the paper prototype should be adjustable so people using it can redesign on the fly e.g. post-its 
    - **Video prototypes**
        - Mackay et al.s method
            - making a physical mock-up of the product 
            - filming an actor interacting with the model as if it was fully functional 
            - the products display dynamics simulated in a animation program 
        - Vertelneys method 
            - a video image is superimposed onto computer graphics 
            - actions are captured against a blue screen and superimposed of the video image onto a premodelled 3D environment
        - restricted by the talent of the production team not the budget         

## In practice 
- When evaluating a prototype one should act as the computer and one should take notes
    - Videotaping can also be useful 
- Some sort of narrative is required 
- Give the user some task they need to finis and some context
    - It is important to let the user interact with the prototype 
- Lo-fi prototype is an essential part of participatory design 
    - People cannot always understand formal models 
    - People can be directly involved in the prototype designs 
- When prototyping the following **questions** are important to consider
    - *Who is the prototype aimed at?*
    - *What is the designer trying to achieve with the prototype?*
    - *What stage of the project are things at and what is the context for the use of the prototype?*
    - *What technologies (hi-fi or lo-fi) are appropriate?*
    
___    
- Prototype **trade-offs**
    - High-quality graphics and animation can be used to create convincing and exciting prototypes *but* they might lead to premature commitment to design decisions 
    - Detailed but special-purposed prototypes help to answer specific questions about the design *but* building a meaningful prototype for each issue is expensive
    - Realistic prototypes increase the validity of user data, *but* they may postpone testing, or require construction of throw away prototypes
    - Iterative refinement of an implementation enables continual testing and feedback, *but* may discourage consideration of radical transformations.

___
- Prototyping is used throughout the design process
- **Requirement animation** is a term used to describe the use of prototyping to illustrate requirements
- **Rapid prototyping** ('throw-it-away' prototyping) uses software such as PowerPoint or Keynote to illustrate concepts 
- **Use-case prototyping** is a 'polished' video, that is produced to a wider audience and to software and hardware development teams whose job it is to bring the product into existence

## Metaphors in design 
- A **metaphor** can generally be seen as taking concepts from one domain and applying them to another 
    - A metaphor in design takes part of the real world understanding and uses it (**Blends**)
- Principles for good metaphor design 
    - **Integration:** Has to do with cohesion and not mixing metaphors 
        - manipulate the whole blend maintaining the web of relationsships
        - the blend has its own structure and needs to be consistently maintained 
    - **Unpacking** people should be able to unpack the blend and understand where the inputs come from and how they work
        - reflection and evaluation can achieve this 
        - designers should only have thing in blend for a good reason
    - **Topology** Different spaces have similar topology 
        - is about how the concepts are organized and structured 
    - **Analysis** When undertaking the analysis the designer should concentrate on getting appropriate features and concepts
        - evaluating how people interpret the metaphor
    - **Design** At the design level designers should consider how to represent objects and actions 
        - they do not need to be realistic e.g. names of menus

## Physical design 
- **Physical design** is concerned with
    - how things are going to work and the feel of the product 
    - taking abstract representations and translating them into physical devices 
- Physical design is about

    - clarifying and presenting the allocation of functions and knowledge between people and devices
- There are three components to physical design 
    - **Operational design** is concerned with specifying how everything works and how the content is structured and stored
    - **Representational design** is concerned with fixing on colors, shapes, sizes and information layout
        - It is concerned with style and aesthetics
    - **Interaction design** in this context is concerned with the allocation of function to humans or technology and structuring sequences of interactions 
___

- A **design language** consists of the following 
    - A set of **design elements** such as the use of color, styles and types of buttons, sliders and other widgets 
    - Some **principles of composition** 
        - Rules for putting them together
    - Collections of **qualifying situation** - context and how the affect use 
    

- Rheinfrank and Evenson method for design includes developing a design language through
    - **Characterization** the process of describing existing assumptions and any preexisting design languages
    - **Re-registration** the creation of a new assumption set through exploring trends and needs through field research 
    - **Development and registration** using storyboards, prototypes and other envisioning techniques
    - **Evaluation** of the reactions to the design
    - **Evolution** of the language over time

## Consolidation
- From a practical POV consolidation is the most difficult part of Contextual design 
- Contextual design require that we bring all the different models together 
- During the consolidation process we are looking for ways which the the system can be redefined 
- The output for the consolidation process is many different consolidation models such as a series of consolidated sequence models 
- Instead of constructing formal models you should instead
    - Review the work models created 
        - If different people have created different models they should each step through their models with the rest of the team
        - The best place to start is with flow models, then sequence models, using artefact, physical and cultural models to fill out the more process-oriented data.
    - Look for common problems, bottlenecks and any effective workarounds 
    - Brainstorm possible solutions

### Using flow models to review roles
- The flow model is the best starting point for getting an overview of the work practice and peoples roles within it 
- To consider how a redesign interactive system could better support the work, you should look at the following
    - Recognize that people switch roles and that different people might share the same role
        - most people has more than one role at work
        - consistent interfaces is a good idea 
    - Minimize overloaded roles where people have many different conflicting tasks that it is impossible to carry any of their roles effectively
        - part of these roles are good places to consider partial or full automation or relocating tasks between roles 
        - when changing parts of peoples jobs it is a good idea to consult the client 
    - Avoid role isolation where roles are so segregated that information sharing and communication becomes problematic
        - make sure that artefacts that communicate between roles have sufficient information
    - As a sanity check review the ideas for change against the fundamental purpose of the customer organization 

### Sequence models as sources of ideas for redesign
- The first question should be whether the activity depicted in the sequence diagram could be automated all together 
- Given that an activity should be kept then consider 
    - Can the intent be achieved in a more effective way, perhaps with part-automation?
    - Are all current steps necessary?
    - The different ways which the different role holders undertake the activity in the sequence
        - some versions might be more effective than others
    - All subsidiary intents
        - make sure the redesign idea still satisfy them
    - Eliminate all unnecessary steps and breakdowns
    - The new design ideas should not create more work 
        - walk through the process to check that this is not the case

### Reconsidering artefacts
- The artefact models gives you clues to where in the process the current artefacts doesn't work well 
- Aspects to consider when reviewing the roles of artefacts in the new design
    - If the redesign intention is to put artefacts online, consider how any informal communications they support will happen in the new system
    - Can some elements of the artefact be completed automatically? 
    - Does the presentation of the artefact highlight the most important aspects in some way 
        - If so how could it be carried to an online version 

### Constrains and values from the cultural model 
- Look for areas where sensitive use of interactive systems can alleviate irritations 
    - often caused by communication problems which a careful system design can approve 
- Be especially careful of deeply held values
- Be aware of the impact of the proposed on power relationships between the stakeholders

### Clues for redesign from the physical model 
- Do not waste time attempting to design an online equivalent of parts of the physical environment which already support work well.
- Consider the impact of proposed changes on the way tasks are currently supported by physical features.
- Look for breakdowns which can be fixed by the new system.
- Consider how the physical environment might be modified - although this is ultimately much more a decision for the customer organization, you should make recommendations where they seem justified.

## Developing a vision 
- The **vision** is a mixture of sketch and text that encapsulates the main points of the new system 
    - It is the main creative part in the CD process
- At this stages you should aim to 
    - Review the ideas that have been captured in the informal consolidation process against the models of work and any other contextual data you have collected.
    - Establish how the client wants to project an image of itself through the information system
        - The cultural model can be helpful here
    - Outline the vision
        - the main functionality of the new interactive
    - Involve all the members of the design team, if you have one, in the creation of the agreed vision 
- After this we define how the new design works in detail, this is done through the following means 
    - Capturing strands of interaction in storyboards
    - Developing a structure for the system in the user environment design 
    - Paper prototyping with customers 

## Constructing a storyboard in Contextual Design
- In CD, storyboards are used to ground the vision of the new system in work practice to make sure it fits with the larger system 
    - Show how real tasks work in a graphical way 
- CD storyboards has a wider perspective than regular storyboards, they incorporate significant interactions both between customers and also between customers and other artefacts
    - More like a cartoon strip 
- Steps to constructing a CD storyboard 
    - Step 1
        - Identify the key tasks which the new system support and choose one
        - Review the work models and affinity diagram for any issues relevant to the tasks
    - Step 2
        - Produce a detailed redesign for the task using text and/or diagrams 
            - do not attempt the storyboard just jet
            - sequence diagrams can be useful
        - Consider alternative options and select that which deals with the intent for the tasks and associated issues most effectively 
    - Step 3
        - Step the redesign through the original sequence models and check that all intents are fulfilled   
    - Step 4
        - Sketch the storyboard, one frame for each step in the task sequence
        - Include interaction between people, automatic and manual steps as well as interactions with the system
        - This is not the place to consider the details of user interface design
            - where screens are include just use a rough sketch 
    - Step 5
        - Repeat for all key tasks 
        - Review the storyboard with the rest of the design team 

## The user environment design (UED)
- To turn what we have so far in the process into a design we need to create a user environment design (UED).
- The **UED** is a customer-centered, high-level design of the interactive application that we intent to implement. 
    - It is customer-centered because it does not use computing notation only plain English with a few boxes and connecting arrows. 
    - should be understood by a non expert who will use the finished system 
- The UED is used to 
    - guide the detailed design of the interactive application 
    - the design of the user interface 
    - the management of the development process 
- A UED is required because a storyboard themselves are not sufficient in themselves to this 


### UED basics 
- The UED is a diagram that represents the structure of the new design. It is:
    - Based on the underlying structure and sequence of the work itself
    - A shared resource for designers, developers and project managers 
    - Constructed before the user interface design takes place 
    - Similar in some ways to a website navigation map 
- A UED comprises a number of **focus areas** 
- The focus areas are particular places in the system that support particular parts of the work
    - collects functions and work object into one place 
    - does not include functions that only affect the UI
    - should be named with a simple active phrase that is the **purpose statement**
    - is not equivalent to a software module
![sample_focus](img/sample_focus.png)
- Each focus area comprises primarily:
    - **A purpose statement:** a brief plain English statement describing the purpose
    - **Functions:** these enable the user to work 
        - described with a short phrasee 
        - invoke either by the user or automatically by the system
        - shown by bullet points in the UED
    - **Objects:** are thing that the user sees and manipulates in the focus area
        - might include people or artefacts
    - **Links:** are pointers to other related focus areas 
- Other areas that a focus area might or might not need are
    - **Constrains:** in the implementation on the focus areas
        - e.g. speed and reliability 
    - **Issues:** which might include ideas for the UI, unresolved problems and so on 
    - **Hidden focus areas** parts of the system the user knows about but doesn't have to interact with 
        - represented by dotted lines
    - **Roles**, a list the users whom you expect to use this focus area

### UED from storyboards
- The UED is constructed from storyboards 
    - is a representation of all storyboards into one single unified structure 
    - constructed by walking thorough all storyboard frame by frame, identifying new focus areas and adding functions links and so forth to existing focus areas 
- When constructing the UED from storyboards the following is worth checking
    - Overlapping focus areas
    - Unnecessary functions
    - Function areas that contain only links - remove them
    - Focus areas that contain more than one task - simplify them.

### Checking the UED with a walkthrough 
- When having created the UED it needs to be checked
    - Ideally a group activity (or getting a friend to listen to you walking through it)
- Things to check for
    - Are focus areas coherent?
    - Support one task in overall activity.
    - Check for focus areas without an obvious purpose.
    - Avoid single-function focus areas.

### UED and the user interface design
- Interface design from the perspective of contextual design, is primarily concerned with the accessing the functions identified in the focus areas of the UED
- The contents of the focus area in the UED might translate into one screen or window but not necessarily
- When designing the details of interactions remember that different roles might use the same focus area, so design for the skills and tasks of each role
- You should review your design against the issues captured in the affinity diagram and work models.

## Paper prototyping 
- Contextual design uses **paper prototyping** as the mean of communicating with people and bringing them into the design process
- A paper prototype essentially animates the interface design 
    - It is constructed using a sheet of paper or better card of the right size to represent the screen and the main elements are permanently displayed 
    - Post-Its acetate transparencies and similar removable media are used to mimic transient or movable windows, dropdown menus and so forth or to try out alternative layouts
- In contextual design paper prototypes are employed in the following ways:
    - Having people walk through their existing tasks 
         - the designer makes the prototype "work"
    - Probing people for the reasons why they take particular actions with the prototype, and what they expect to happen.
    - Taking notes of reactions, suggestions and confusions
        - having a second designer involved in the session helps here.
     - At an early stage of design, involving people in the review of the general scope and organization of functionality.
    - As the design develops, checking the effectiveness of a more complete and detailed interface design.
    - Using a "replay" of the prototyping session to communicate user responses to the rest of the design team.

# Prototypes as Filters, Prototypes as Manifestations of Design Ideas
- Two fundamental aspects of prototypes form the basis of the prototyping framework: 
    1. Prototyping are for traversing a design space, leading to the creation of meaningful knowledge about the final design as envisioned in the process of design
    2. Prototypes are purposefully formed manifestations of design ideas.
- The initial set of design aspects that a prototype might exhibit is called **filtering dimensions**
    - The designer ignores unnecessary details of the prototype
    - The decision on what to filter out is based on the purpose of prototyping
![principles_prototyping](img/principles_prototyping.png)
 

## Prototypes as filters
- Prototypes are helpful as much in what they do not include as in what they do.
    - Helps the designer make decisions about certain aspects without thinking about the others 
- They are used for discovering problems and new solutions 
- Prototypes in this context are a means of **generative and evaluative discovery**
- The most efficient prototype is the most incomplete one that still filters the qualities the designer wants to examine and explore.    

## Prototypes as Manifestations of Design Ideas
- The **Externalization of thought** gives rise to new perceptual and cognitive operations that allow for reflection, critique, and iteration.
- The designer performs creates and envisions an idea by realizing the idea in some physical manifestations
    - can take almost any form, shape and appearance, based on the material choice 
- Working with the design of a digital artifact is not limited by the choice of material such as a physical one 

## Anatomy of Prototypes
- The proposed anatomy includes
    1. **Filtering dimensions**
        - **Appearance:** is physical properties of the design 
            - may include forms, colors, textures, sizes, weights, shapes and proportional relationships
            - not restricted to visual appearance since the characteristics can be sensed by touch as well as sight 
        - **Data:** is the information architecture and the data model of the design
            - may include 
                - the size of data
                - the number of letters to be shown in each label
                - the amount of visible and invisible data on screen
                - the semantic organization of the contents, the ways of labeling and naming
                - the levels of privacy of data
                - the types of information
        - **Functionality:** is the functions that can be performed by the design
            - designers may determine preferred functionalities and scenarios associated with using different functions.
        - **Interactivity:** is the ways in which people interact with each part of the system 
            - it may include feedback, input behaviors, operation behaviors, and output behaviors
        - **Spatial structure:** is how each component in the system is connected with each other
            - it may include considerations of laying out interface or information elements in an interactive space.
    2. **Manifestation dimensions**
        - includes materials, resolution, and scope
        - even though they are influenced by the filtering dimension the choice of manifestation dimensions involves various issues such as resources, cost, and user perception in the use of a prototype.
        - influence how well a prototype performs as an informing tool in the design process.

# User centered agile methods
- **Customers** are people who derive value from the system
    - users' management
    - the purchaser
    - possibly the it department
- **Stakeholders** are people in the development organization who depend on the system being correct

## Common Agile Methods
- The two most common agile methods are scrum and XP 

### Scrum 
- **The product owner** is the customer representative on a scrum team
    - often played by the product manager or a marketing person
    - defines what the product stakeholders need
    - must find out what stakeholders and end users actually need
    - must reconcile requirements and communicate them to the team 
    - must protect the team from problems caused by stakeholders changing their minds 
    - requirements can only be changed at the start of a sprint 
    
    
- **The team** 
    - All members of the scrum team are equally committed and responsible for producing results 
    - Only team members get a say in team decisions 
        - Others get to express an opinion but not make a decision 
    - The team is guided by a key individual, the **ScrumMaster**
        - manages the process 
        - runs the daily stand up meeting 
        - runs the planning meetings to determine the product and sprint backlog 
        - maybe played by a new role or the product manager if the product manager does not play the role of the product owner
        
        
- **The sprint** 
    - Development in scrum is organized into sprints 
    - They are usually a month long but some teams use shorter sprints 
    - They start by choosing user stories to implement in that sprint 
        - only as many stories that can be completed by the end of that sprint are chosen 
        - task are defined to implement the stories and the team work starts 
        - stories and tasks are usually tracked using sticky notes or maybe online using a tool
    - Scrum defines several project management tools to track the process doing a spring 
        - e.g. burndown charts maintained by the scrum master
    - Each day starts with a stand-up meeting for the whole team to share process and plan activities of the day 
        - is a standing meeting to encourage a short meeting 
        - a common format is for each team member to report on what they have done, what they plan to do this day, and what is getting in their way.
    - At the end of the sprint 
        - there is a review of the sprint with the product owner and stakeholders
            - is a short walkthrough no more than 4 hours 
        - the teams direction is reevaluated, stories are reprioritized, and new stories maybe written to respond to feedback
        - the team also reflects on its own processes, identifies problems, and discusses changes to make their process for the next sprint.
        
        
- **The backlog** 
    - Scrum projects start with product and sprint backlogs 
    - Each backlog item is a unit of work to be completed within a sprint 
    - User stories are the primary backlog items 
    - At the beginning of the project, the team develops and priorities the product backlog 
    - Scrum does not define how the product owner knows what user stories to put on the backlog.
    - The product owner is expected to capture individual system requirements as stories
    - The design will emerge from the code through iterative development 
    - In practice, the backlog items are often derived from a Product Requirements Document (PRD) written by the product owner using traditional methods
    
    
- **User stories** 
    - Each element of the desired solution is described in a **user story**
    - The story is written on a index card or a sticky note
        - Spreadsheets or other special purpose tools are often used in practice 
    - The story is not intended to be a full description of the feature
        - captures enough about that feature to remind everyone what that feature is 
    - When a coder starts working on a feature their first steps will be to talk to the product owner and find out the exact UI to implement, if it requires one
    - Should be small enough to complete within one sprint
    - Common format for writing stories is *“As a [user role], I want [a feature] so that I can [achieve some goal].”* 

### Extreme programming - XP
- **The customer** is a real and experienced end user can be a full team member
    - is empowered to decide what is or is not useful in the product
    - They determine the priority of features and what is or is not useful in the product 
    - define the detailed system design, look, layout, and interaction to meet their needs.
    - in most cases the customer is a role played by one person or a team of several internal people, usually including the product owner or project manager


- **The release planning game** 
    - The XP customer arrives with user story cards describing everything that they think is needed in the next release of the product
    - Developers then estimate the implementation time required for each story (their velocity).
    - The result of the release planning game is a rough estimate of when the project will be completed
    - XP emphasizes tracking how much work a team can do in one sprint and using the measure to limit the work they commit to
    - Stories can be added or removed to change delivery date 


- **Iterations** 
    - In XP sprints are referred to as iterations
    - Each iteration starts with iteration planning and ends with customer acceptance
    - The length of the iterations are typically two weeks
    - Every user story for an iteration must be entirely completed within that iteration, including passing all unit and customer acceptance tests


- **Iteration planning**
    - Each iteration starts with an iteration planning session
    - This is a face-to-face meeting in which the team selects the stories to be implemented during an iteration.
    - The XP customer chooses the most important to provide value 
    - The team uses their velocity to decide how many stories to include
    - Once the stories are selected their are broken down into tasks that can be done by different people on the team 
    - Any rework has to be done during the iterations
        - Rework includes bug fixes and functionality changes 
    - Rework can be handled in one of the following to ways
        - Treat it as overhead e.g. reserving 20% of their points 
        - Plan rework using story cards - gets its own card and work estimate
    - The XP customer is responsible for prioritizing rework against implementing new stories


- **Acceptance testing**
    - XP culture is strongly test-driven, and it values automated testing highly
    - Teams are expected to do nightly builts and run their entire testing suite daily 
    - Teams want customer acceptance tests to be automatable and, ideally, defined along with the user stories
        - can show whether the system meets the technical requirements of the customer 
    - Quality Assurance groups test the whole product at the end of each iteration, often as team members
        - Usability test can also be performed at the end


- **Development practices** 
    - XP defines a number of development practices intended to support rapid development with little documentation 
        - Pair programming, test-driven development, collective code ownership, and the nightly build are a few of these.
    - Because XP and Scrum address slightly different problems they are often used together 
        - Scrum provides the overall project management
        - XP provides more detailed guidance on running development.

## Agile culture
- Agile methods often think the team as members with no special skills where one person can be exchanged for another
- UX people should be recognized as full team members and educate the rest of the team on what UX is about and what is needed to design a reasonable UI
- UX people can help understand their users and their users need. 
    - can provide the methods and expertise for building detailed user knowledge
___    
- The problem of generating user stories and creating a design is not covered by the agile methods and can be done by UX designers
    - can be done in phase 0 before the development is started by the UX designer
    - UX designer should contributed by understanding users, representing users work and tasks and designing the appropriate information to meet the users needs
    - The goal is to design and validate the core of the proposed system so that user stories can be written with confidence that these stories define a successful product.
___    
- The UX designer can be used to understand the users needs when creating a new feature
    - The developers and UX designers should have a close working relationship 
    - It should be a norm to have daily discussions with one developer on how to implement the UI for a specific feature
    - In addition, UX professionals may also need to let go of some of their own documentation, relying more on sketches and discussions and less on fully-rendered UI designs.
___    _
- The UX work should be broken up in the same way as sprints
    - may have be done an iteration ahead of the same function  
- Putting off bug fixing is called **technical debt**
- Putting off usability testing or losing control of design coherence across the UI because the different parts are being iterated independently is just another form of debt called **design debt**.
    - To limit this the ongoing UX activities must be to maintain an architectural view of the entire product 
___    
- A reason for getting for having small sprints getting is feedback on each sprint as it is developed.
    - This ensures that the project stays on track 
- It is the UX team who has the skills to get the real user feedback
    - Real validation of work requires that the team test the product with users in their own work context, working through examples of their own work tasks
    - If UX designers can design, test, and iterate a screen design directly with users faster than it can be developed and iterated in code, this is just another way of reducing the lag.

## Best Practices for Integrating UX with Agile
- **Get user feedback from real users in context**
    - **Users are not good at articulating what they do**
        - This is the problem knowledge 
        - The requirements they give are inaccurate and incomplete 
    - **Users want to be helpful** 
        - Users tend not to tell if they have big problems with the system 
        - Users often only report minor inconveniences with the system 
    - **Users are not available as team members**
        - Real users have real jobs
        - Users has not the time
    - **Surrogate users aren't.**
        - Surrogate users are never a good stand-in for real end-users.
        - Hiring users to be on the development team often takes the developers side  
    - **Marketing methods do not collect design data**
        - Marketing methods are good at collecting sales points and market requirements; they do not collect design data.
        - An user-centered agile process must include real user research 
    - **Online data collection is incomplete** 
        - can report on what the users did but not why, or what they were trying to accomplish, or what the larger work task was.
        - they team will not discover what other opportunities they do not know about


- **A phase 0 to define system scope and structure**
    - It is important to get a Big Picture Up Front (BPUF) 
        - Aviods misunderstandings with the user
    - An effective agile process will make room for some high-level, up-front user research and high-level design, tested and iterated 
        - often called **phase-0** 
        - grounds the agile customer team in the real world practise practice, culture, goals, strategies, and issues. 
        - defines the groundwork for the user experience
        - helps writing user-stories in confidence
  
  
- **UI design done one iteration ahead**
    - It is important to give UX designers room to breather by letting the start on a component design one or two sprints before the coding starts
        - the UX team gets the whole of the sprint to do their UI design and user testing
        - testing can and should be iterative with users         
    - The following sprint the UX team member can communicate the detailed UI to the developer


- **Validation Done One Iteration Behind**
    - Successful agile teams have found it effective to do their their real testing one sprint after the implementation
    - The sprint completes and the internal Quality Assurance (QA) or User Acceptance Testing (UAT) processes confirm that they have correctly implemented the stories of that sprint. 
    - After the implementation sprint the UX  people can take the baselevel to customers, walk through real examples with it, and bring back problems and fixes to be addressed in the following sprint.


- **Parallel UX stream**
    - The UX team is always concerned with the overall consistency 
        - can be hard to maintain a higher focus 
    - To make sure that the higher focus is not lost separate a UX stream can be created
        - is not part of any agile development stream 
        - it runs in parallel as syncs with the agile team at major points
        - can do additional user research, design, and iteration without being tied to the particular schedule and deliverables of any one development stream.
        - maintains the whole system coherence of design
    
    
- **Programmer/Designer holiday**
    - The programmers holiday dedicates an iteration to code clean up
        - Each developer choose what they work on 
        - Can also be useful to a UX team where they can look at the overall coherence of the UI 
    

- **Architectural spikes for difficult issues**
    - Architectural spikes allows the development team to address a difficult or risky development problem
        - might devote a whole sprint to studying a technical problem prototyping alternative solutions or running load tests
        - happen early in development
    - Not all project risk come from implementation concerns e.g. challenging UI problems
    - When the team identifies a UI risk it is appropriate to devote a iteration to addresing that risk 
        - The can use the time to brainstorm alternatives, mock them up, and test them with users to verify that they actually work as implemented
    
    
- **UX as a full team member**
    - Success UX team members act like whole members
    - The UX person should ideally be there 
    - UX team members can also act use the teams resources      
        

## Structure of a User-Centered Agile Process

### Phase 0: Project Definition 
- Starting with a phase 0, gives the team a change to organize themselves
- Phase 0 is for creating a Big Picture Up Front 
    - becomes the basis for iteration during sprints
- The first step is to see how users approach their work
    - The best practice for understanding work practice is Contextual Inquire 
    - To do this a contextual interview is used
    - Interpretation sessions can be used to give the team a way to deal the field data 
        - In an interpretation session the team review each interviews by going through the interviews notes in detail
        - The interviewer retells the story of the interview from beginning to end, in order. 
        - The rest of the team captures key information which has design relevance.
        - Individual observations are written down to a list a used later in the affinity diagram
        - The work practice of the user is captured in work models 
        - By the end all the important points about the user has been written down and are ready to use 
        - The notes are used by the agile customer team to understand the customer
- In this phase we use the following techniques
    - Contextual inquire and affinity diagrams
    - Work modeling
    - Personas
    - Visioning
    - Storyboards
    - User environment design
        - gives the customer team a way to represent the structure of the system as experienced by the user
        - does not show detail of appearance 
            - like a floor plan for software systems
        - Focus areas are places 
            - has a purpose and the work it is intended to support
            - provides functions which describes what the system shows and does for the user in that place
            - provides access to other Focus Areas 
    - Paper prototyping
        - is done two times in the agile process
            - a rough sketch in phase 0 - two rounds
            - right before the implementation           

### The release planning session
- User stories are written on card 
    - should be a simple high-level description of a function or behavior 
    - Complex functions are spread across several stories so each is small enough to fit in a sprint
        - This helps with flexibility and predictability
- When the story cards are written the team estimates the cost of implementing them
    - maybe done in "ideal programmer days"
    - experienced teams often find it more convenient to assign "points" to stories
        - points has no real world counterpart
        - process is measure in a teams velocity which is the amount of points they can implement in a single sprint
        - works best for experienced teams
- Once stories are written and estimated, they can be organized into sprints
    - The sprints should be implementable, which means that if it relies on other features not included in the sprint they should be implemented first
    - The team should allow a buffer of 20% following sprint 1 or plan regular programmers holidays
    - The first two sprints should be planned carefully with the exact stories chosen 
        - The following can be planned more roughly
    - The release is a trade-off between function and date
        - Since the sprints takes a fixed known time, it can be estimate how many sprint will be needed to deliver the critical function to the release.
        - The UX designers should be prepared to argue for which feature to include in the release to support the users' work practice 
            - It is the UX designer who understand the real impact of the proposed system on the end-users
        - The release data can be adjusted by moving stories between releases 
            - More stories should not be jammed into sprints
        - As the users knows more about the system and the developers knows more about the users the plan will change
        - It must be understood that product marketing might need a specific feature to compete in the marketplace       

### Running sprints
![sprint_planning](img/sprint_planning.png)
- To give the UX designers time to develop the final UI and test them with users the best practice is to work one sprint ahead of developers and test one sprint behind
    - In sprint 1, the UI people develop and test the final screen appearance for the user story
    - In sprint 2, they work with development to implement it
    - In sprint 3, they test the implementation to ensure that nothing is lost in translation
- The first sprint might not provide any actual user value, but it may put all the building blocks in place for the project       

#### Sprint planning
- The team meets to plan the work for the sprint
    - Starts with a set of stories identified during release planning
    - The specific stories for the current sprint needs to be determined
- Most teams will write **task cards** for each of the current stories in the current implementation
    - can represent implementation tasks, such as, writing code and designing databases
    - can represent UX tasks: user test, paper prototypes
    - represent other tasks such as documentation and QA 
    - The UX team need task cards for the stories in the last, the current and the next sprints
        - The cards for the current specify UI design taks: doing low_level layout and visual design to support developers
        - They write tasks card for the next sprint to start the UI design on those stories: taking the high-level design from vision, storyboards, and UED and to design the final layout
            - If paper prototypes was not created in phase 0 they may need to or three rounds of paper prototypes first which is tested with users during the current sprint.
        - The also writes card for the work completed in the last sprint to test working code to make ensure that the final implementation actually works for people

#### Working with development
- The UX team has three jobs running in parallel reflected in the three types of task cards
- The UX team need to support developers by providing detailed UI design and consult them about the detailed behavior and look
    - It is up to the UX team to be tightly tied to the developer team
- The UX runs customer visits throughout the iteration to bring real user feedback into the development process
    - done by running a customer interview that combine contextual inquiry and paper prototyping

#### Sprint interview
- Before the interview the UX team identify which elements of work which is done need user feedback
    - Should be focused on the new parts of the UI which the users are not familiar with 
    - The UX team also looks at the stories to be implemented in the next sprint and finds the trouble spots 
        - The UI design for these stories will need to be started and tested with users before the next sprint starts
        - To test these stories, the UX team develops the final, fully-rendered visual design and then chooses how to prototype it with users
            - Online tools and developers can be used if paper prototypes are not good enough
- The interview runs as follows 
    - The interviewer introduces themselves and their focus to the user
        - tell the users which elements they want to test and the tasks that the prototype supports
        - get an overview of the users work
        - plans how to run the interview and what they will cover
    - If the team needs to test a new design, the interviewer goes into paper prototyping mode
    - The interviewer walk through the users real world examples
        - discus issues with they system and how they might be fixed
    - If the team needs to understand a work tasks the interview goes into CI mode when the user does the task

#### Interpretation session 
- The data from a interview is analysed in a interpretation session 
    - The UX team and any interested part of the full project team goes through event of the interview 
    - The team captures notes and also validations
- After the interpretation session the team evaluates the results of the session
    - Uses these findings to produce a revised prototype
- Notes identifying issues in the last sprint are more complicated
    - The team redesign the interface to solve the problem and writes a story card which represents that fix
    - The new story card will be priorities in the next sprint planning session 
- A person on the team has the responsibility for finding users to these sessions    

## Structuring projects
- If the UX practitioner are dropped into a project that is already using agile methodology the designer should move toward greater user involvement using the following series of steps 
    1. **Do User Test in the Users' Workplace**
        - If the team is not currently working with actual end-users set up field visits to the users
            - Arrange the visit so it is possible to see them in a real work context
            - Initially use interviews to prototype UIs 
        - If the team currently brings users in to test product iteration
            - Move some of the sessions out in the field. 
            - Use customer visits to run hybrid interviews 
                - where a part of the visit support detailed design of the new UI
                - where the other part provides user test and feedback of the completed implementations 
    2. **Work Towards Design a Sprint Ahead**
        - Start working on the UI design for the next sprint
        - If it is a design your less familiar with make a rough paper prototype and test it
        - Spend part of the interview getting feedback on paper prototype of stories for the next sprint and part getting feedback on the stories for the current sprint
    3. **Start Gathering Work Practice Data**  
        - Organization collect two kinds of data
            - Marketing requirements of wish lists and desires 
            - The results of usability test
        - Once field interviews are happening the team can start collecting actual work practice data
            - Start with observation notes and sequence models for the tasks of interest to the project
                - They are straightforward to capture and can quickly consolidated
            - Use the initial part of the interview as more general CI to capture the data
    4. **Use a Project Break to Step Back**
        - After collecting enough work data the UX team should step back and evaluating the overall direction
            - Done after in a few iteration 
            - Should be done between release or a programmer's holiday
        - Consolidate work data, building models and affinity diagram
        - Build a UED of the system as implemented to data and extend it using stories not implemented yet
        - Find the places the project is failing to address important work issues
            - The UX team tests the fixes with users and writes new stories


- Contextual interviews and sequence model can help gain understand of the new features when extending an existing product
    - During interpretation sessions the team capture sequence models and notes for affinity 
    - Leads to a understanding of how the task is done as a whole 
    - New interfaces are designed and tested using paper prototypes
    - User stories are written and based on the vision and final prototypes
    - Done when there are a small number of tasks to support


- If it is a major new release the UX team needs the full range of research
