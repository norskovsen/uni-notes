## Problem 11
### Delopgave a

```r
dpois(0,1/30*60*4)*100
```

### Delopgave b

- $W = $ it is a weekday
- $W^c = $ it is weekend
- $N$ no mail received for one hour
- $P(W) = \frac{5}{7}$
- $P(W^c) = \frac{2}{7}$  
Find P(W|N)

```r
pw = 5/7
pwc = 2/7

# P(N|W)
pnw = dpois(0, 1/6*60)

# P(N|W^c)
pnwc = dpois(0, 1/30*60)

# P(N) (Using law of total probability)
pn = pnw*pw + pnwc * pwc

# P(W|N) (Using Bayers rule)
pwn = (pnw*pw)/pn
print(pwn*100)
```

## Øvelse 14
### Delopgave a
$$
\left[\overline{X}- t_{\frac{\alpha}{2},n-1}\frac{S}{\sqrt{n}} , \overline{X}+ t_{\frac{\alpha}{2},n-1} \frac{S}{\sqrt{n}}\right]
$$

```r
m = 35.8
s2 = 12.5
s = sqrt(s2)
a = 0.1
n = 36
lower = m-qt(1-a/2,n-1)*s/sqrt(n)
upper = m+qt(1-a/2,n-1)*s/sqrt(n)
print(paste("90% konfidensinterval",sprintf("[%f, %f]",lower,upper), sep=" "))
a= 0.05
lower = m-qt(1-a/2,n-1)*s/sqrt(n)
upper = m+qt(1-a/2,n-1)*s/sqrt(n)
print(paste("95% konfidensinterval",sprintf("[%f, %f]",lower,upper), sep=" "))
a = 0.01
lower = m-qt(1-a/2,n-1)*s/sqrt(n)
upper = m+qt(1-a/2,n-1)*s/sqrt(n)
print(paste("99% konfidensinterval",sprintf("[%f, %f]",lower,upper), sep=" "))
```

### Delopgave b
$$
    \left[\frac{(n-1)S^2}{\chi^2_{\frac{\alpha}{2},n-1}} , \frac{(n-1)S^2}{\chi^2_{1-\frac{\alpha}{2},n-1}} \right]
$$    

```r
m = 35.8
s2 = 12.5
s = sqrt(s2)
a = 0.1
n = 36
lower = ((n-1)*s2)/(qchisq(1-a/2,n-1))
upper = ((n-1)*s2)/(qchisq(a/2,n-1))
print(paste("90% konfidensinterval",sprintf("[%f, %f]",lower,upper), sep=" "))
a = 0.05
lower = ((n-1)*s2)/(qchisq(1-a/2,n-1))
upper = ((n-1)*s2)/(qchisq(a/2,n-1))
print(paste("95% konfidensinterval",sprintf("[%f, %f]",lower,upper), sep=" "))
a = 0.01
lower = ((n-1)*s2)/(qchisq(1-a/2,n-1))
upper = ((n-1)*s2)/(qchisq(a/2,n-1))
print(paste("99% konfidensinterval",sprintf("[%f, %f]",lower,upper), sep=" "))
```

## Øvelse 16
### Delopgave a
$$
    \left[\overline{X}- z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}} , \overline{X}+ z_{\frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}\right]
$$

```r
xs = c(16.34,18.57,18.22,16.94,15.98,15.23,17.22,16.54,17.54)
m = mean(xs)
s2 = 1
s = sqrt(s2)
a = 0.1
n = length(xs)
lower = m - qnorm(1-a/2)*s/sqrt(n)
upper = m + qnorm(1-a/2)*s/sqrt(n)
print(sprintf("[%f, %f]",lower,upper))
```

$\mu_0$ er ikke inkluderet i intervallet 

### Delopgave b
$$
    W = \frac{\overline X - \mu_0}{\sigma/\sqrt n}
$$
Accept region:
$$
    |W| \leq z_{\frac \alpha 2}
$$

```r
m0 = 16
W = (m-m0)/(s/sqrt(n))
print(sprintf("|%f| ≤ %f",W,qnorm(1-a/2)))
```

Vi kan udfra dette afvise vores $h_0$ med et signifikansniveau på 10%
